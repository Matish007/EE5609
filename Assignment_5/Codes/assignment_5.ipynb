```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y_1 = 3*x+2 # straight line 1
y_2 = 2*x+(1/2) # straighe line 2
plt.plot(x, y_1, '-r', label='-6x+2y-4=0')
plt.plot(x, y_2, '-b', label='-2x+y-1/2=0')
plt.title('Graph of two straight lines')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='best')
plt.grid()
plt.savefig('straightlines.png', dpi=300)
plt.show()
```
